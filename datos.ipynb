```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('usuarios_con_edad.csv')

# Obtener el rango de edades
edad_minima = df['edad'].min()
edad_maxima = df['edad'].max()

print(f"El rango de edad en el archivo es de {edad_minima} a {edad_maxima} años.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df = pd.read_csv('usuarios_con_edad.csv')

# Crear el histograma de frecuencias de edad
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='edad', bins=20, kde=True, color='skyblue')
plt.title('Histograma de Frecuencias de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
